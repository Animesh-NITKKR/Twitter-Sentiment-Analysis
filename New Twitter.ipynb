## Defining a function to extract data from tweet object

```python
def extract_tweet_attributes(messy_tweet):
    # creating an empty list
    
    tweet_list = []
    # Now looping through messy tweets
    for tweet in messy_tweet:
        tweet_id = tweet.id # unique integer idetifier for tweet
        text = tweet.text # get text from every tweet
        favorite_count = tweet.favorite_count
        retweet_count = tweet.retweet_count
        created_at = tweet.created_at # utc time tweet created
        source = tweet.source # utility used to post tweet
        reply_to_status = tweet.in_reply_to_status_id # if reply int of orginal tweet id
        reply_to_user = tweet.in_reply_to_screen_name # if reply original tweetes screenname
        retweets = tweet.retweet_count # number of times this tweet retweeted
        favorites = tweet.favorite_count # number of time this tweet liked
        # append attributes to list
        
        tweet_list.append({'tweet_id':tweet_id, 
                          'text':text, 
                          'favorite_count':favorite_count,
                          'retweet_count':retweet_count,
                          'created_at':created_at, 
                          'source':source, 
                          'reply_to_status':reply_to_status, 
                          'reply_to_user':reply_to_user,
                          'retweets':retweets,
                          'favorites':favorites})
        df = pd.DataFrame(tweet_list, columns=['tweet_id',
                                           'text',
                                           'favorite_count',
                                           'retweet_count',
                                           'created_at',
                                           'source',
                                           'reply_to_status',
                                           'reply_to_user',
                                           'retweets',
                                           'favorites'])
    return df
```

# removing user handles

```python
df['text_without_handle'] = df['text'].replace(to_replace = '(@[\w]+)',value ='',regex = True)
df['text_without_handle'] = df['text_without_handle'].replace(to_replace =('RT'), value ='',regex = True)
df
```

# removing any kind of links from tweet

```python
df['text_without_handle'] = df['text_without_handle'].str.replace('((www\.[\s]+)|(https?://[^\s]+))','\0',regex=True)
df
```

# Removing special char,numbers and punctuation

```python
df['text_without_handle'] = df['text_without_handle'].str.replace("[^a-zA-Z]+", " ")
df.head()
```

```python
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
import re
from nltk.corpus import stopwords
```

```python
df['text_without_handle'] = df['text_without_handle'].str.lower()
```

```python
df['text_without_handle'] = df['text_without_handle'].apply(lambda x: [item for item in word_tokenize(x)])
```

```python
stop = stopwords.words('english')
```

```python
df['text_without_handle'] = df['text_without_handle'].apply(lambda x: [item for item in x if item not in stop])
```

# applying lemmatizatio


```python

w_tokenizer = nltk.tokenize.WhitespaceTokenizer()
lemmatizer = nltk.stem.WordNetLemmatizer()

def lemmatized_text(text):
    return [lemmatizer.lemmatize(w) for w in w_tokenizer.tokenize(text)]


    
```

```python
df['text_without_handle'] = df['text_without_handle'].apply(lemmatized_text)
```

```python
def rejoin_words(row):
    my_list = row['text_without_handle']
    joined_words = ( " ".join(my_list))
    return joined_words

df['text_without_handle'] = df.apply(rejoin_words, axis=1)
```
